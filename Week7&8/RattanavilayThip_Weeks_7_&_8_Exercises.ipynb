```python
# Chapter 7: Task 1 : #Filter out missing data
# Filtering where Q1: GOING OUT? is NaN
# I think this question is relevant, if one needs to carry out the study
# So the values with missing answer to this question should be removed
df_candy_2017_filtered = df_candy_2017.dropna(subset=['Q1: GOING OUT?'])
df_candy_2017_filtered

# Rows Reduced to 2350 rows from 2460 rows
```

```python
# Chapter 7 : Task 2: 
# a. Replace values: 
# b. Transform data using either mapping or a function
import re

# This function returns value as -1, if the data is not numeric
def CheckNum(x):
    if re.match('^[0-9]*$', x):     
        return x
    else:
        return -1

# Checking Unique Values before change
print("Unique Values Before - ", df_candy_2017['Q3: AGE'].unique())

# Changing value and column type as NaN if the 
df_candy_2017['Q3: AGE'] = df_candy_2017['Q3: AGE'].astype(str).apply(lambda x: CheckNum(x))

# Converting column back as Int
df_candy_2017['Q3: AGE'] = df_candy_2017['Q3: AGE'].astype(int)

# Check unique values after transformation to see if the values have been replaced
print("Unique Values After - ", df_candy_2017['Q3: AGE'].unique())
```

```python
# Chapter 10: Task 1 : Grouping with Functions - using count() function 
# to group them based on Department and Accession Year
df_group_by = df_metro.filter(
    items=[
        'Department','Object Name','AccessionYear'
    ]
)
# count() adds up all the Objects which have similar Department and Accession Year
df_group_by = df_group_by.groupby(['Department','AccessionYear']).count()
print(df_group_by)
```

```python
# Chapter 10: Task 2 : Grouping with Index Levels
df_group_by = df_metro.filter(
    items=[
        'Department','Object Name','AccessionYear'
    ]
)
#df_group_by
df_group_by = df_group_by.set_index(["Department","AccessionYear"],inplace=False)
df_group_by.sort_index(level=[0])
df_grp = df_group_by.groupby(by=["Department","AccessionYear"]).count()
df_grp
```

```python
# Chapter 10: Task 3 : Cross Tab
df_group_by = df_metro.filter(
    items=[
        'Department','Object Name','AccessionYear'
    ]
)
df_cross = pd.crosstab(df_group_by.Department,df_group_by.AccessionYear)
df_cross
```
