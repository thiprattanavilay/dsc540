```python
#import libraries
import urllib.request, urllib.parse, urllib.error
import requests
import pandas as pd
from bs4 import BeautifulSoup
import ssl
import re
import json
import matplotlib.pyplot as plt
```

# Connecting to an API/Pulling in the Data and Cleaning/Formatting

```python
#called the api from api.covid19api.com
api_url = "https://api.covid19api.com/summary"
headers = {'Content-Type': 'application/json'}
response = requests.get(api_url, headers=headers)
response_json = response.json()
```

```python
#Created a variable that would read the json specifically using countries
countries = (response_json['Countries'])
```

```python
#Created the dataframe
df = pd.DataFrame(countries)
df
```

```python
#Slug has everything in lowercase as those are proper nouns so I'm going to capitalize the first letter 
df['Slug'] = df['Slug'].str.capitalize()   
df 
```

```python
#Want to split up the date/time column into separate columns one for date and one for time
df['Dates'] = pd.to_datetime(df['Date']).dt.date
df['Time'] = pd.to_datetime(df['Date']).dt.time
df
```

```python
#Now I can drop the original column Date
df.drop(columns=['Date'])
```

```python
df.to_csv('covid_clean_data3.csv')
```
