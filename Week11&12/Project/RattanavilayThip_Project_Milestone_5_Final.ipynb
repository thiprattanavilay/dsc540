## Merging all 3 Data and Storing in a Database/Visualizing Data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load data from Milestone 2

```python
#Reading in csv created from milestone 2
df_covid1 = pd.read_csv("covid_clean_data1.csv")
df_covid1.head()
```

```python
#For some reason the unnamed column came back so need to drop that
df_covid1 = df_covid1.drop(columns = ['Unnamed: 0'])
df_covid1.head()
```

```python
#Going to re-name the country column
df_covid1.rename(columns={'Country,Other':'Country'}, inplace = True)
df_covid1.head()
```

### Load data from Milestone 3

```python
#Reading in csv created from milestone 3
df_covid2 = pd.read_csv("covid_clean_data2.csv")
df_covid2.head()
```

```python
#Removing # column
df_covid2 = df_covid2.drop(columns = ['#'])
df_covid2.head()
```

```python
#Renaming country column
df_covid2.rename(columns={'Country,Other':'Country'}, inplace = True)
df_covid2.head()
```

```python
#Didn't need the world information so dropped first row
df_covid2 = df_covid2.drop([0], axis=0)
df_covid2.head()
```

```python
#My other datasets were set to alphabetical order by country. 
df_covid2.sort_values(['Country'], axis = 0, ascending = True, inplace = True)
df_covid2 = df_covid2.reset_index(drop = True)
df_covid2.head()
```

### Load data from Milestone 3

```python
#Reading in csv created from milestone 4
df_covid3 = pd.read_csv("covid_clean_data3.csv")
df_covid3.head()
```

```python
#Removing unnamed column
df_covid3 = df_covid3.drop(columns = ['Unnamed: 0', 'Date'])
df_covid3.head()
```

## Merging dataframes

```python
#Merging my first two dataframes. Using combine first so df_covid1 is my source of truth
merge = df_covid1.combine_first(df_covid2)
merge
```

```python
#Merging my third dataframe with the others and creating df_total
df_total = merge.combine_first(df_covid3)
df_total
```

```python
#Not sure why my cleaning didn't save, but changed NaN into 0's
df_total.fillna(0,inplace=True)
df_total
```

## Save dataframe to CSV

```python
#saved df to csv
df_total.to_csv('covid_total.csv')
```

## Load DB dataset using SQLLite

### - Combined all three dataset into one DB file -

```python
# import library
import sqlite3
```

```python
# connecting to covid.db
conn = sqlite3.connect("covid.db")
```

```python
cur=conn.cursor()
```

```python
# select function from table and execute
cur.execute('SELECT * from covid_data')
```

```python
# execute and print rows
for row in cur.execute('SELECT * FROM covid_data'):
    print(row)
```

```python
#Created a dataframe out of my sqlite table
df_covid19 = pd.read_sql_query("SELECT * FROM covid_data", conn)
df_covid19
```

## 5 Visualization using Matplotlib

```python
#Looking at New Cases of Covid 19 by New Deaths caused by Covid 19
x=df_covid19.NewCases
y=df_covid19.NewDeaths
plt.scatter(x,y,color='green')
plt.xticks(rotation=90)
plt.grid(True)
plt.title("New Cases of Covid against New Deaths")
plt.xlabel("New Cases of Covid")
plt.ylabel("New Deaths")

plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Created a historgram of total cases against total deaths. Cannot figure out how to make the x-axis more ledgible 
plt.hist([df_covid19.TotalCases, df_covid19.TotalDeaths])
plt.xticks(rotation=90)
plt.title("Total Cases against Total Deaths")

#save image
plt.savefig('plot2.png', dpi=300, bbox_inches='tight')
```

```python
#Horizontal bar chart of total cases
df_covid19['TotalCases'].value_counts()[:20].plot(kind='barh')

#save image
plt.savefig('plot3.png', dpi=300, bbox_inches='tight')
```

```python
#Bar chart of total cases by continent
ac = df_covid19[['Continent','TotalCases']]
ac = ac.Continent.value_counts().sort_index() # group and count
ac = pd.DataFrame(ac) # convert series to dataframe again
ac = ac.reset_index() # reset index so columns match
ac.columns = ['Continent', 'TotalCases'] # re-name the columns back
plt.bar(ac.Continent, ac.TotalCases)
plt.xticks(rotation=90, fontsize=15)
plt.xlabel("Continents")
plt.ylabel("Total Cases")
plt.show()

#save image
plt.savefig('plot4.png', dpi=300, bbox_inches='tight')
```

```python
#Bar chart of total tests by continent
tt = df_covid19[['Continent','TotalTests']]
tt = tt.Continent.value_counts().sort_index() # group and count
tt = pd.DataFrame(tt) # convert series to dataframe again
tt = tt.reset_index() # reset index so columns match
tt.columns = ['Continent', 'TotalTests'] # re-name the columns back
plt.bar(tt.Continent, tt.TotalTests, color='red')
plt.xticks(rotation=90, fontsize=15)
plt.xlabel("Continents")
plt.ylabel("Total Tests")
plt.show()

#save image
plt.savefig('plot5.png', dpi=300, bbox_inches='tight')
```
